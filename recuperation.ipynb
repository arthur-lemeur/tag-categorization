```python
from stackapi import StackAPI
import pandas as pd

APP_KEY = "rl_dfB7mZZDJY4Efab78EKrdtjiq"

SITE = StackAPI('stackoverflow', key=APP_KEY)
SITE.page_size = 100  # max autorisé par l'API
SITE.max_pages = 500  # pour arriver à 50000
SITE.sleep_between_requests = 0.1

questions = SITE.fetch(
    'questions',
    filter='withbody',
    sort='creation',
    min=1,
    tagged='',
)

questions_with_tags = [
    {
        'Title': q.get('title'),
        'Body': q.get('body'),
        'Tags': q.get('tags'),
        'Id': q.get('question_id'),
        'Score': q.get('score'),
        'ViewCount': q.get('view_count'),
        'FavoriteCount': q.get('favorite_count'),
        'AnswerCount': q.get('answer_count')
    }
    for q in questions['items']
    if q.get('tags') and len(q['tags']) >= 1
]

print(f"Questions récupérées : {len(questions_with_tags)}")
```

```python
df = pd.DataFrame(questions_with_tags)
df['Tags'] = df['Tags'].apply(lambda tags: ''.join(f'<{t}>' for t in tags))
```
